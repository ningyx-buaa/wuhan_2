```python
#新闻数据的读取
news_id = list(news['news_id'])
news_text = list(news['content'])
news_title = list(news['title'])
news_time = list(news['time'])
news_url = list(news['url'])
news_customer = list(news['customer'])
news_userview = list(news['userview'])
news_location = list(news['location'])
news_country = list(news['country_label'])
news_content = list(news['content_label'])
news_data = {}
for i in range(len(news_id)):
    temp = {}
    if type(news_text[i]) != str and math.isnan(news_text[i]):
         news_text[i] = ""
    temp['text'] = news_text[i]
    if type(news_title[i]) != str and math.isnan(news_title[i]):
         news_title[i] = ""
    temp['title'] = news_title[i]
    temp['time'] = news_time[i]
    temp['url'] = news_url[i]
    if type(news_customer[i]) != str and math.isnan(news_customer[i]):
         news_customer[i] = ""
    temp['customer'] = news_customer[i]
    if type(news_userview[i]) != str and math.isnan(news_userview[i]):
         news_userview[i] = 0
    temp['userview'] = news_userview[i]
    if type(news_location[i]) != str and math.isnan(news_location[i]):
         news_location[i] = ""
    temp['location'] = news_location[i]
    temp['country_l'] = news_country[i]
    temp['content_l'] = news_content[i]
    news_data[news_id[i]] = temp
```
